```python
import numpy as np
import tensorflow as tf
import PIL as pil

```

```python
#DL model
x=100
input=(100,100,3)
model=tf.keras.Sequential([
                           tf.keras.layers.Conv2D(x,(3,3),activation='relu',input_shape=input),
                           tf.keras.layers.MaxPooling2D(2,2),  #Can be avg
                           tf.keras.layers.Conv2D(x,(3,3),activation='relu'),
                           tf.keras.layers.MaxPooling2D(2,2),
                           tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(x)),
                           tf.keras.layers.Flatten(),
                           tf.keras.layers.Dense(512, activation='relu'),
                           tf.keras.layers.Dense(1, activation='sigmoid')
])
```
